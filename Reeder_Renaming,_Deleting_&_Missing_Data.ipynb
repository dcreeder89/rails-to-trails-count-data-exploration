## Import pandas and data into dataframe

```python
import pandas as pd
```

```python
# Save data from excel file as dataframe
filename = '/content/drive/MyDrive/Coding Dojo/02 Week 2: Pandas for Data Manipulation/Core Assignments/RailsToTrails_National Count Data_week.xlsx'
bike_df = pd.read_excel(filename)
bike_df.info()
bike_df.head()
```

## 1. Delete unnecessary columns

```python
# Delete unnecessary columns
bike_df = bike_df.drop(columns=['Unnamed: 7', 'Unnamed: 8', 'Unnamed: 9'])
# Display dataframe head to check that columns are gone
bike_df.head()
```

## 2. Rename the Unnamed:6 column to percent_change

```python
# Rename column
bike_df = bike_df.rename(columns={'Unnamed: 6':'percent_change'})
# Display dataframe to check column ID changed
bike_df.head()
```

## 3. How many missing values are in the dataset? Which columns are they in?

```python
# Add up all null values in the dataset and display sums for each column
bike_df.isna().sum()
```

## 4. Fill all missing values in the '2021 Counts' column with 0

```python
# Fill all NaN values in '2021 Counts' column with 0
bike_df['2021 Counts'].fillna(0, inplace=True)
# Check null values count per column to check the change
print(bike_df.isna().sum())
# Check dataframe head to check that the values were actually changed to 0
bike_df.head()
```

## 5. Bonus

> There is a challenging but common issue in the name for the 2019 counts column. Can you find it? If so, rename the 2019 counts (31 counters) column to "counts_2019" (a more Pythonic column name)

```python
# Change column name of ' 2019 counts (31 counters)' to counts_2019 using dictionary
bike_df = bike_df.rename(columns={' 2019 counts (31 counters)':'counts_2019'})
# Display dataframe to check column ID change
bike_df.head()
```

> The issue is there is a space added before 2019, Python needs all characters present in the string, including spaces
